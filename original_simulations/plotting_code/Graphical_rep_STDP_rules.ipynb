```python
# Import packages.
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
import seaborn as sns
from sympy import symbols, Eq, solve
from sympy import Matrix
import scipy as sp
```

## Learning window of classical Hebbian STDP

```python
# Plot mean EE weight over time for different timescales 'eta'.
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.9, rc={"lines.linewidth":3.3})

tauSTDP = 20 # or 200ms

t_k = np.linspace(1,100,10000)
t_j = np.linspace(1,100,10000)


W_prepost = np.exp(-t_k/tauSTDP)

W_postpre = - np.exp(-t_j/tauSTDP)


plt.axhline(y=0,linestyle='--', color='k',linewidth=2)
plt.axvline(x=0,linestyle='--', color='k',linewidth=2)

plt.plot(-t_j,W_prepost)
plt.plot(t_k,W_postpre)

plt.xlabel('$t_{pre} - t_{post}$')
plt.ylabel('$\Delta j_{ee}/\eta_{ee}$')

plt.xlim((-100,100))
plt.xticks((-100,0,100))

sns.despine()

plt.show()

fig.savefig("../../PlotsReview_BNs/Fig9-learning-windows/learning_window_classic_Hebb.pdf", bbox_inches="tight")

```

## Learning window of weight--dependent Hebbian STDP

```python
# Plot mean EE weight over time for different timescales 'eta'.
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.9, rc={"lines.linewidth":3.3})

tauSTDP = 20 # or 200ms

Jmax = 30

Jee = 25

t_k = np.linspace(1,100,10000)
t_j = np.linspace(1,100,10000)

W_prepost = Jmax * np.exp(-t_k/tauSTDP)

W_postpre = - Jee * np.exp(-t_j/tauSTDP)


plt.axhline(y=0,linestyle='--', color='k',linewidth=2)
plt.axvline(x=0,linestyle='--', color='k',linewidth=2)

plt.plot(-t_j,W_prepost)
plt.plot(t_k,W_postpre)

plt.xlabel('$t_{pre} - t_{post}$')
plt.ylabel('$\Delta j_{ee}/\eta_{ee}$')

plt.xlim((-100,100))
plt.xticks((-100,0,100))

sns.despine()

plt.show()

fig.savefig("../../PlotsReview_BNs/Fig9-learning-windows/learning_window_modified_Hebb.pdf", bbox_inches="tight")
```

## Learning window of iSTDP

```python
# Plot mean EE weight over time for different timescales 'eta'.
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.9, rc={"lines.linewidth":3.3})

tauSTDP = 20 # or 200ms

rho0 = 0.010 # 10 Hz

alpha = rho0 * 2 * tauSTDP

t_k = np.linspace(1,100,10000)
t_j = np.linspace(1,100,10000)

W_prepost = np.exp(-t_k/tauSTDP) 

W_postpre = np.exp(-t_j/tauSTDP) - alpha


plt.plot(-t_j,W_prepost)#/np.abs(np.max(W_prepost)))
plt.plot(t_k,W_postpre)#/np.abs(np.max(W_postpre)))

plt.axhline(y=0,linestyle='--', color='k',linewidth=2)
plt.axvline(x=0,linestyle='--', color='k',linewidth=2)

plt.xlabel('$t_{pre} - t_{post}$')
plt.ylabel('$\Delta j_{ei}/\eta_{ei}$')

plt.xlim((-100,100))
plt.xticks((-100,0,100))

sns.despine()

plt.show()

fig.savefig("../../PlotsReview_BNs/Fig9-learning-windows/learning_window_istdp.pdf", bbox_inches="tight")


```
