```python
# Import packages.
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
import seaborn as sns
# Load the .mat file.
mat_contents=sio.loadmat('./timescales_analysis_hebbian_ee_soft.mat')
```

```python
c = mat_contents['c'].squeeze()
dt = mat_contents['dt'].squeeze()
dtRate = mat_contents['dtRate'].squeeze()
eRateT = mat_contents['eRateT']
iRateT = mat_contents['iRateT']
eta_ee_soft_vector = mat_contents['eta_ee_soft_vector'].squeeze()
print(eta_ee_soft_vector)
Ie_current = mat_contents['Ie_current_eta']
Ii_current = mat_contents['Ii_current_eta']
Ix_current = mat_contents['Ix_current_eta']
Jm = mat_contents['Jm']
Jmax_ee_soft = mat_contents['Jmax_ee_soft'].squeeze()
Jxm = mat_contents['Jxm']
mC = mat_contents['mC']
mean_Jee_recorded = mat_contents['mean_Jee_recorded']
mR = mat_contents['mR']
mV = mat_contents['mV']
N = mat_contents['N'].squeeze()
rx = mat_contents['rx'].squeeze()
seed = mat_contents['seed'].squeeze()
T = mat_contents['T'].squeeze()
T1 = mat_contents['T1'].squeeze()
T2 = mat_contents['T2'].squeeze()
tauSTDP = mat_contents['tauSTDP'].squeeze()
W = mat_contents['W']
winsize = mat_contents['winsize'].squeeze()
Wx = mat_contents['Wx']
time_axis = np.linspace(0,T/1000,len(mean_Jee_recorded[0,:]))
```

```python
# Plot mean EE weight over time for different timescales 'eta'.
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.6, rc={"lines.linewidth":2.3})

# for i in range(len(mean_Jee_recorded[:,0])-2):
#     plt.plot(time_axis,mean_Jee_recorded[i,:])

line1, = plt.plot(time_axis,mean_Jee_recorded[0,:],label="$\eta=%.4f$" %eta_ee_soft_vector[0])
line2, = plt.plot(time_axis,mean_Jee_recorded[1,:],label="$\eta=%.3f$" %eta_ee_soft_vector[1])
line3, = plt.plot(time_axis,mean_Jee_recorded[2,:],label="$\eta=%.2f$" %eta_ee_soft_vector[2])
line4, = plt.plot(time_axis,mean_Jee_recorded[3,:],label="$\eta=%.1f$" %eta_ee_soft_vector[3])

plt.axhline(y=Jmax_ee_soft*np.sqrt(N),color='darkslategray',linestyle='--')

plt.xlabel(r'Time $(s)$')
plt.ylabel(r'Exc weight')

leg = plt.legend(handles=[line1,line2,line3,line4,line5],loc='upper left',fontsize=16,frameon='none',markerscale=1)
leg.get_frame().set_linewidth(0.0)

ax = plt.gca().add_artist(leg)

# leg1 = plt.legend(handles=[line5],loc='center left',fontsize=16,frameon='none',markerscale=1)
# leg1.get_frame().set_linewidth(0.0)

plt.ylim((23,47))
plt.xlim((-1,T/1000))

plt.xticks((0,T/1000))
plt.yticks((25,35,45))

sns.despine()
plt.show()

fig.savefig("./PlotsReview_BNs/timescale_hebbian_soft.pdf", bbox_inches="tight")
```
