```python
# Import packages.
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
import seaborn as sns
# Load the .mat file.
mat_contents=sio.loadmat('./timescale_analysis_eiSTDP.mat')
```

```python
c = mat_contents['c'].squeeze()
dt = mat_contents['dt'].squeeze()
dtRate = mat_contents['dtRate'].squeeze()
eRateT = mat_contents['eRateT']
iRateT = mat_contents['iRateT']
eta_ei_iSTDP_vector = mat_contents['eta_ei_iSTDP_vector'].squeeze() * (-200/np.sqrt(N))
print(eta_ei_iSTDP_vector)
Ie_current = mat_contents['Ie_current_eta']
Ii_current = mat_contents['Ii_current_eta']
Ix_current = mat_contents['Ix_current_eta']
Jm = mat_contents['Jm']
Jxm = mat_contents['Jxm']
mC = mat_contents['mC']
mean_Jei_recorded = mat_contents['mean_Jei_recorded']
mR = mat_contents['mR']
mV = mat_contents['mV']
N = mat_contents['N'].squeeze()
rx = mat_contents['rx'].squeeze()
seed = mat_contents['seed'].squeeze()
T = mat_contents['T'].squeeze()
T1 = mat_contents['T1'].squeeze()
T2 = mat_contents['T2'].squeeze()
tauSTDP = mat_contents['tauSTDP'].squeeze()
W = mat_contents['W']
winsize = mat_contents['winsize'].squeeze()
Wx = mat_contents['Wx']
time_axis = np.linspace(0,T/1000,len(mean_Jei_recorded[0,:]))

J_ei_theory = -127
```

```python
# Plot mean EI weight over time for different timescales 'eta'.
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.6, rc={"lines.linewidth":3.3})

# for i in range(len(mean_Jee_recorded[:,0])-2):
#     plt.plot(time_axis,mean_Jee_recorded[i,:])

line2, = plt.plot(time_axis,mean_Jei_recorded[1,:],label="$\eta=%.3f$" %eta_ei_iSTDP_vector[1])
line3, = plt.plot(time_axis,mean_Jei_recorded[2,:],label="$\eta=%.2f$" %eta_ei_iSTDP_vector[2]
        ,color='tab:pink')
line4, = plt.plot(time_axis,mean_Jei_recorded[3,:],label="$\eta=%.1f$" %eta_ei_iSTDP_vector[3]
        ,color='tab:red')

plt.axhline(y=J_ei_theory,color='green',linestyle='--')

line1, = plt.plot(time_axis,mean_Jei_recorded[0,:],label="$\eta=%.4f$" %eta_ei_iSTDP_vector[0])

plt.xlabel(r'Time $(s)$')
plt.ylabel(r'Inh weight')

leg = plt.legend(handles=[line1,line2],loc='center',fontsize=16,frameon='none',markerscale=1)
leg.get_frame().set_linewidth(0.0)

ax = plt.gca().add_artist(leg)

leg1 = plt.legend(handles=[line3,line4,line5],loc='center right',fontsize=16,frameon='none',markerscale=1)
leg1.get_frame().set_linewidth(0.0)


plt.ylim((-650,-50))
plt.xlim((-2,T/1000))

plt.xticks((0,T/1000))
plt.yticks((-600,-350,-100))

sns.despine()
plt.show()

fig.savefig("./PlotsReview_BNs/timescale_eiSTDP.pdf", bbox_inches="tight")
```
